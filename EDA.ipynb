```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
from pathlib import Path
```

```python
# Load the data
file_path = Path('./DataTables/heart_2020_cleaned.csv')
df = pd.read_csv(file_path)
df_raw = pd.read_csv(file_path)
```

```python
categorical = ["Smoking", "AlcoholDrinking", "Stroke", "DiffWalking", "Sex", "AgeCategory", "Race", "Diabetic", "PhysicalActivity", "GenHealth", "Asthma", "KidneyDisease", "SkinCancer"]
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
for x in df.columns:
    if df[x].dtype=='object':
        df[x]=enc.fit_transform(df[x])
df.info()
```

```python
df.describe()
```

```python
profile = ProfileReport(df, title = 'Heart disease',correlations = None, interactions = None, html = {'style':{'full_width' : True}})
profile.to_widgets()
```
